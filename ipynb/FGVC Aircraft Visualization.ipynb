```python
import torch
import torchvision
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset= torchvision.datasets.FGVCAircraft('.', split= 'train', annotation_level= 'variant', download = True)
```

```python
iterator = iter(dataset)
x, y = next(iter(dataset))
print(len(dataset))
```

```python
iterator = iter(dataset)
num_rows = 8
num_cols = 8
fig, axs = plt.subplots(8, 8, figsize=(12, 9))
for row in range(num_rows):
    
    for col in range(num_cols):
        x, y = next(iterator)
        ax = axs[row][col]
        ax.imshow(x)
        ax.set_title("Label {}".format(y))
        ax.axis("off")
        
plt.show()
```
