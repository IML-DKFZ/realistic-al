```python
from pathlib import Path

import pandas as pd
from IPython.display import Markdown, display

# base_path = Path("/home/c817h/network/Cluster-Experiments/activelearning/cifar10_imb/full_data").resolve()
base_path = Path("/home/c817h/network/Cluster-Experiments/activelearning/sweep/isic2019").resolve()
paths = base_path.glob("*/*.csv")

modes = set(map(lambda path: str(path.stem), paths))
# print(modes)

for mode in sorted(modes):
    if "val_w_acc" in mode:
        paths = list(base_path.glob(f"*/{mode}.csv"))

        if len(paths) == 0:
            print(f"No files found for mode {mode}")
            continue

        data = (
            pd.concat([pd.read_csv(path, index_col=0) for path in paths])
            .sort_values("Experiment")
            .reset_index(drop=True)
        )
        data_vis = data
        data_vis = data[data["Experiment"].str.contains("balancsamp")]
        display(Markdown(f"## {mode}"))
        display(data_vis.style.background_gradient()) #.format(precision=2))
```
